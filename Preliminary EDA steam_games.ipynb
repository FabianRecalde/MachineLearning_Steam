```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# Cargando archivos en el datagrame
steam_games = pd.read_json("steam_games.json", lines = True) # "lines = true" significa que el archivo tiene varios objetos json separados
steam_games
```

```python
# Limpia todas las líneas que no tienen ningún dato.
steam_games.dropna(how='all', inplace=True)

steam_games
```

```python
# Reemplazando 'Free To Play' con 0 en la columna 'precio'
steam_games['price'] = steam_games['price'].replace('Free To Play', 0)
steam_games['price'] = steam_games['price'].replace('Free to Play', 0)

# Convertir la columna 'precio' en un tipo numérico
steam_games['price'] = pd.to_numeric(steam_games['price'], errors='coerce')
```
